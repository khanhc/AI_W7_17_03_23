```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
url = "/data_crab.csv"
data = pd.read_csv(url)
data.head()

data.hist()
plt.show()
```

```python
sns.pairplot(data)
correlation = data.corr()
print(correlation)
```

```python
data.plot(kind='scatter',x='CW',y='CL',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['CW']), y = data['CL'])
print('gia tri intercep:', model.intercept_)
print('gia tri coffient:', model.coef_)
print('do chinh xac:', model.score(X = pd.DataFrame(data['CW']), y = data['CL']))
```

```python
data.plot(kind='scatter', x='CW', y='CL', figsize=(9,9), color='black', xlim=(0,65))
y_pred = model.predict(X=pd.DataFrame(data['CW']))
plt.plot(data['CW'], y_pred, color='blue')
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['CW'], data['CW']**2]).T
model.fit(X,y = data['CL'])
print('Intercept la:', model.intercept_)
print('coeffient la:', model.coef_)
print('do chinh xac la:', model.score(X,y = data['CL']))
```

```python
range = np.arange(15,55,1)
X_range = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter', x='CW', y='CL', figsize=(9,9), color='black', xlim=(0,65))
plt.plot(range, y_pred, color='blue')
```

```python

model=linear_model.LinearRegression()
X = pd.DataFrame([data['CW'], data['CW']**2, data['CW']**3, data['CW']**4, data['CW']**5, data['CW']**6, data['CW']**7, data['CW']**8, data['CW']**9, data['CW']**10]).T
model.fit(X,y = data['CL'])
print('Intercept la:', model.intercept_)
print('coeffient la:', model.coef_)
print('do chinh xac la:', model.score(X,y = data['CL']))
```

```python
range = np.arange(15,55,1)
X_range = pd.DataFrame([range, range**2, range**3, range**4, range**5, range**6, range**7, range**8, range**9, range**10]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter', x='CW', y='CL', figsize=(9,9), color='black', xlim=(0,65))
plt.plot(range, y_pred, color='blue')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
url = "/passenger.csv"
data = pd.read_csv(url)
data.head()

data.hist()
plt.show()
```

```python
sns.pairplot(data)
correlation = data.corr()
print(correlation)
```

```python
data.plot(kind='scatter',x='Dollars (Y)',y='Miles(X)',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Dollars (Y)']), y = data['Miles(X)'])
print('gia tri intercep:', model.intercept_)
print('gia tri coffient:', model.coef_)
print('do chinh xac:', model.score(X = pd.DataFrame(data['Dollars (Y)']), y = data['Miles(X)']))
```

```python
data.plot(kind='scatter', x='Dollars (Y)', y='Miles(X)', figsize=(9,9), color='black', xlim=(0,7500))
y_pred = model.predict(X=pd.DataFrame(data['Dollars (Y)']))
plt.plot(data['Dollars (Y)'], y_pred, color='blue')
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['Dollars (Y)'], data['Dollars (Y)']**2]).T
model.fit(X,y = data['Miles(X)'])
print('Intercept la:', model.intercept_)
print('coeffient la:', model.coef_)
print('do chinh xac la:', model.score(X,y = data['Miles(X)']))
```

```python
range = np.arange(1500,7000,1)
X_range = pd.DataFrame([range, range**2]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter', x='Dollars (Y)', y='Miles(X)', figsize=(9,9), color='black', xlim=(0,7500))
plt.plot(range, y_pred, color='blue')
```

```python
model=linear_model.LinearRegression()
X = pd.DataFrame([data['Dollars (Y)'], data['Dollars (Y)']**2, data['Dollars (Y)']**3, data['Dollars (Y)']**4, data['Dollars (Y)']**5, data['Dollars (Y)']**6, data['Dollars (Y)']**7, data['Dollars (Y)']**8, data['Dollars (Y)']**9, data['Dollars (Y)']**10]).T
model.fit(X,y = data['Miles(X)'])
print('Intercept la:', model.intercept_)
print('coeffient la:', model.coef_)
print('do chinh xac la:', model.score(X,y = data['Miles(X)']))
```

```python
range = np.arange(1500,7000,50)
X_range = pd.DataFrame([range, range**2, range**3, range**4, range**5, range**6, range**7, range**8, range**9, range**10]).T
y_pred = model.predict(X=X_range)
data.plot(kind='scatter', x='Dollars (Y)', y='Miles(X)', figsize=(9,9), color='black', xlim=(0,7500))
plt.plot(range, y_pred, color='blue')
```
